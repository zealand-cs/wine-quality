```python
#Define labels for wine quality - That is true, if the quality is high. Otherwise false (low quality)
highQLabels=[False,True]
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

#Define the quality label for each wine
y_true = [True, False, True, True, False, True, True, False, True, True, False, True, False, False, False, True, True, False, True, True]

#Assume that the quality for each wine has been predicted as follows
y_pred = [False, False, True, True, False, True,False, False, True, True, False, True, True, False, False, True, True, False, True, False]

#Establish the confusion matrix
cm=confusion_matrix(y_true, y_pred, labels=highQLabels)
cm
```

```python
#Display the confusion matrix
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(cm, display_labels=highQLabels)
disp.plot()
plt.show()
```

```python
#Calculate the precision score: TP/(FP+TP) = 9/(1+9) = 0.9
from sklearn.metrics import precision_score
precision_score(y_true, y_pred)
```

```python
#Calculate the recall score: TP/(FN+TP) = 9/(3+9) = 0.75
from sklearn.metrics import recall_score
recall_score(y_true, y_pred)
```

```python
#Calculate the accuracy score: (TN+TP)/(TN+FP+TP+FN) = (7+9)/(7+1+9+3) = 0.8
from sklearn.metrics import accuracy_score
accuracy_score(y_true, y_pred)
```

### Assignment 1: What does the performance measures mean?

Explain how the precission, recall and accuracy performance measures can make sense?

What does that mean in our Portuguese wine case?

Add text cells below for your answer.

Precision represents how much 'high quality' wine the model predicted there are vs how much there actually is. A precision of 0,9 would mean that our model identifies 90% of our wine as 'high quality'.

Recall means the proportion of 'high quality' wines our model identifies. A recall of 0,9 here would mean we got 90% of the high-quality wines, but got 10% false negatives.

Accuracy is the overall correctness of the model across multiple parameters, meaning both high and low quality wines. 0,9 here would mean it classified 90% of both high and low quality wines correctly.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

mlp_clf = MLPClassifier(hidden_layer_sizes=[5], activation="relu", max_iter=10_000, random_state=42)
pipeline = make_pipeline(StandardScaler(), mlp_clf)
pipeline
```

1. Neural networks like MLP are sensitive to the scale of input features.    Features with larger magnitudes can dominate the learning process. `StandardScaler` standardizes features by removing the mean and scaling to unit variance, helping the model converge faster and perform better.

2. A pipeline bundles the preprocessing steps (like scaling) and the estimator (the classifier) into a single object. This prevents data leakage (by ensuring the same transformations are applied to train and test sets) and makes the code cleaner and easier to manage.

3. `hidden_layer_sizes` defines the number of hidden layers and the number of neurons in each. `[5]` means one hidden layer with 5 neurons.

4. `activation` is the activation function for the hidden layers. `'relu'` (Rectified Linear Unit) helps the network learn non-linear relationships and is generally efficient to train.

5. `max_iter` is the maximum number of iterations (epochs) the solver will run to find the optimal weights. If the model doesn't converge before this limit, it stops.

6. `random_state` ensures reproducibility. Using a fixed number (42) means the random initialization of weights will be the same every time you run the code.

7. The total layers include the input layer, hidden layers, and the output layer. We can check this using the `n_layers_` attribute after fitting, or calculate it from the configuration.
